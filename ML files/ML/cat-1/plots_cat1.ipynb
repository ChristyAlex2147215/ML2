```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv("/content/processed-mushrooms - mushrooms.csv")
```

```python
df
```

```python
dff=df[["class","cap-shape","cap-surface","cap-color","bruises","odor"]].head(700)
```

#BOX Ploat

```python
import matplotlib.pyplot as plt
```

```python
ph=dff["odor"].head(20)
```

```python
phh=ph.astype('float')
fig,ax = plt.subplots()
ax.boxplot(phh,vert=0)
plt.show()
```

#Bar Ploat

```python
classes = dff['class'].head(50)
odor = dff['odor'].head(50)
 
# Figure Size
fig = plt.figure(figsize =(10,15))
 
# Horizontal Bar Plot
plt.bar(classes, odor)
 
# Show Plot
plt.show()
```

```python
import seaborn as sns
```

#Dist *`Ploat`*

```python
plot=sns.distplot(dff["cap-shape"],kde=False,color="b")
```

#Strip ploat

```python
sns.stripplot(x="cap-shape",y="cap-surface",data=dff) 
```

#Pie Ploat

```python
from matplotlib import pyplot as plt
import numpy as np
```

```python
cars=[x for x in dff['bruises'].head(5)]
data=[c for c in dff['odor'].head(5)]
print(cars)
print(data)
```

```python

# Creating plot
fig = plt.figure(figsize =(10, 7))
plt.pie(data, labels = cars)
 
# show plot
plt.show()
```
